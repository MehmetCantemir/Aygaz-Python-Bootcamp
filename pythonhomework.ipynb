```python
import random
oyun_secimi = ["taş","kağıt","makas"]

# Oyuncunun taş , kağıt , makas seçimini yaptığı fonksiyon
def oyuncu_secimi():
    while True:
        # Kullanıcıdan giriş alıyoruz
        kullanici_secimi = input("Seçiminizi yapın (taş, kağıt, makas): ").lower()
        # Eğer giriş doğruysa döngüden çıkıyoruz
        if kullanici_secimi in oyun_secimi:
            return kullanici_secimi
        # Eğer yanlışsa uyarı veriyoruz ve döngüye devam ediyoruz
        else:
            print("Geçersiz seçim! Lütfen 'taş', 'kağıt' veya 'makas' girin.")

# Oyunun oynandığı fonksiyon
def tas_kagit_makas_MEHMET_CANTEMIR():
    kullanici_puani=""    # oyuncu puanı
    bilgisayar_puani=""   # bilgisayar puanı
    tur=""                # tur sayısı
    oyun_devam_mi=True    # oyun bitişinde yeni oyuna geçiş durumu

    # Oyun başlangıcı
    while oyun_devam_mi:

      # puanları ve turu sıfırlama
      kullanici_puani=0
      bilgisayar_puani=0
      tur=0

      # Bilgisayar veya oyuncu puanlarının 2 olana kadar oyunu devam ettiren döngü
      while kullanici_puani < 2 and bilgisayar_puani < 2:

        # Oyuncu ve bilgisayar seçimlerini alıp gösterme
        oyuncu_secimi = oyuncu_secimi()
        bilgisayar_secimi = random.choice(oyun_secimi)
        print(f"Bilgisayarın Seçimi: {bilgisayar_secimi}")
        print(f"Kullanıcının Seçimi: {oyuncu_secimi}")


        # Seçimlerin beraberlik durumu
        if oyuncu_secimi == bilgisayar_secimi:
            print("Berabere!")


        # Oyuncunun kazandığı durum
        elif (oyuncu_secimi == "taş" and bilgisayar_secimi == "makas") or \
             (oyuncu_secimi == "kağıt" and bilgisayar_secimi == "taş") or \
             (oyuncu_secimi == "makas" and bilgisayar_secimi == "kağıt"):
             kullanici_puani += 1
             print("Kazandınız!")
             print("Puanınız : "+str(kullanici_puani))
             print("Bilgisayar Puanı :"+str(bilgisayar_puani))


        # Bilgisayarın kazandığı durum
        else:
            bilgisayar_puani += 1
            print("Bilgisayar Kazandı!")
            print("Puanınız : "+str(kullanici_puani))
            print("Bilgisayar Puanı: "+str(bilgisayar_puani))

        # Tur sayısını arttırma ve bilgi verme
        tur += 1
        print(f"Toplam {tur} tur oynadınız.")


        # Oyuncunun veya bilgisayarın puanına göre oyunu bitirme
        if bilgisayar_puani==2 or kullanici_puani==2:
          if bilgisayar_puani==2:
            print("Bilgisayar Kazandı!")
          else:
            print("Kazandınız!")


        # Yeni oyun başlatma veya oyunu bitirme
        if kullanici_puani==2 or bilgisayar_puani==2:
          print("Yeniden Oynamak İster misiniz ?")
          yeniden_oyna = input("Seçiminizi yapın (evet, hayır): ").lower()
          if(yeniden_oyna=="hayır"):
            oyun_devam_mi=False
            print("Oyun Bitti")



# Oyun fonksiyonunun çağırımı.
tas_kagit_makas_MEHMET_CANTEMIR()
```
